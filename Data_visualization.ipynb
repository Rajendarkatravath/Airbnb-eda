1.let's see what hosts (IDs) have the most listings on Airbnb platform and taking advantage of this service.

```python
top_host=df['host_id'].value_counts().head(10)

print(top_host)
```

```python
#coming back to our dataset we can confirm our fidnings with already existing column called 'calculated_host_listings_count'
top_host_check=df['calculated_host_listings_count'].max()


print(top_host_check)
```

```python
#setting figure size for future visualizations
sns.set(rc={'figure.figsize':(12,6)})
sns.set_style('whitegrid')
```

```python
top_host_df=pd.DataFrame(top_host)
top_host_df.reset_index(inplace=True)
top_host_df.rename(columns={'index':'Host_ID', 'host_id':'P_Count'}, inplace=True)

top_host_df
```

```python
sns_palette=sns.color_palette("terrain_r", 7)
viz=sns.barplot(x="Host_ID", y="P_Count", data=top_host_df,
                 palette=sns_palette)
viz.set_title('Hosts with the most listings in NYC')
viz.set_ylabel('Count of listings')
viz.set_xlabel('Host IDs')
```

conclusion:

By this Bar chart we can easily understand the top 10 host_id though the p_count values.we can see that there is a good distribution between top 10 hosts with the most listings.Fisrt host_id 219517861 has more than 300+ listings.The last Host_id in between the top 10 host_id is 7503643 has more than 50 listings.

12.finding the density and distribution of prices for each neighberhood_group

```python
# finding the density and distribution of prices for each neighberhood_group

sub_6=df[df.price > 200]
#using violinplot to showcase density and distribtuion of prices 
sns.violinplot(data=sub_6, x='neighbourhood_group', y='price')
plt.title('Density and distribution of prices for each neighberhood_group')
plt.xlabel('neighbourhood')
plt.ylabel('price')

plt.show()
```

conclusion:

 with a statistical table and a violin plot we can definitely observe a couple of things about distribution of prices for Airbnb in NYC boroughs. First, we can state that Manhattan has the highest range of prices for the listings with $150 price as average observation, followed by Brooklyn with $90 per night. Queens and Staten Island appear to have very similar distributions, Bronx is the cheapest of them all.
